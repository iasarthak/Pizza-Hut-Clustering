## Kmeans

```python
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score

errors=[]
inertia=[]
for cluster in range(2,20):
    kmeans = KMeans(n_clusters = cluster)
    kmeans.fit(dataset_final)
    errors.append(davies_bouldin_score(dataset_final,kmeans.labels_))
    inertia.append(kmeans.inertia_)

frame1 = pd.DataFrame({'Cluster':range(2,20), 'SSE':errors})

plt.figure(figsize=(12,6))
plt.plot(frame1['Cluster'], frame1['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('davies bouldin score')
plt.show()

frame2 = pd.DataFrame({'Cluster':range(2,20), 'SSE':inertia})

plt.figure(figsize=(12,6))
plt.plot(frame2['Cluster'], frame2['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
df=pd.DataFrame()
for i in range(1,11):
    kmeans=KMeans(4)
    kmeans.fit(dataset_final)
    df['Lab'+str(i)]=np.array(kmeans.labels_)
    
df
```
